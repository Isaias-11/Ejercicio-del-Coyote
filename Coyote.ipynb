>Un cuerpo cuadrado de b=3[m] de masa m=2000[kg] que pivota sobre la esquina (punto O). Suponga que la piedra rueda sin deslizar y que hay una fricción en la rueda con µ=0.3 .



# Solución 
 Datos:
<dt> b=3 [m]<dt> m=2 000 [kg] <dt> u=0,3<dt>
<dt> <dt> 1.- Determinamos la inercia en el punto O. <dt>
    $$I_0=\frac{1}{2}m(b^2+b^2)+m\frac{b}{2}=\frac{5}{12}$$
<dt> 2.- Realizamos sumatoria de momentos en O <dt>
    $$\sum M_o=T_o\alpha$$ <dt> $$mg\frac{b}{2}\sin(\theta)=\frac{5}{12}mb^2\alpha$$ <dt>
    De donde tenemos: <dt>
    $$\alpha=\frac{12g\sin(\theta)}{10b}$$ <dt> $$\alpha=\frac{6g}{5b} \sin(\theta)$$ <dt> Determinamos la velocidad angular "w" <dt> $$\alpha=\frac{dw}{dt}$$ <dt> $$\frac{dw}{dt} \frac{d\theta}{d\theta}=\frac{6g}{5b} \sin(\theta)$$ <dt> $$\frac{dw}{d\theta}W=\frac{6g}{5b} \sin(\theta)$$ <dt> $$\int_{0}^{w} w \cdot dw=\frac{6g}{5b}\int_{0}^{\theta} \sin(\theta) \cdot d\theta$$ <dt> $$\frac{w^2}{2}=-\frac{6g}{5b}(\cos(\theta)-1)$$ <dt> $$w^2=\frac{12g}{5b}(1-\cos(\theta))$$ 
    

# DCL: De la piedra
![title](DCL1.jpeg)


4.- Sumatoria en el eje transversal <dt> $$\sum F_{er}=ma_{er}$$ <dt> $$N-mg\cos(\theta)=-m\frac{b}{2}w^2$$ <dt> $$N=mg\cos(\theta)-m\frac{b}{2}w^2$$ <dt> $$N=mg\cos(\theta)-mg\frac{b}{2}\frac{12g}{5b}(1-cos(\theta))$$ <dt> $$N=\frac{mg}{5}(11\cos(\theta)-6)$$ <dt> 5.-Ahora para determinar el anguna, tomamos en cuenta cuando ya no existe contacto entre la piedra y la esquina, esto quiere decir que la normal será igual a cero. <dt> Entonces: $$N=0$$ <dt> $$0=\frac{mg}{5}(11\cos(\theta)-6)$$ <dt> $$\cos(\theta)=\frac{6}{11}$$ <dt> Entonces tenemos que el angulo cuando empieza a caer la piedra es: <dt> $$\theta=56,94^{\circ}$$

```python
#Datos del ejercicio
b=3
m=2000
u=0,3
g=9.81
#Ecuaciones y aálisis
ang=np.arange(0,57,0.1)
W=np.sqrt(((12*g)/(5*b))*(1-np.cos(ang*(np.pi/180))))
#plot
fig=px.line(x=ang,y=W,labels={'x':'Ángulo [grados]','y':'Velocidad angular[rad/s]'},title="Velocidad angular VS Angulo teta")
fig.show()
```
